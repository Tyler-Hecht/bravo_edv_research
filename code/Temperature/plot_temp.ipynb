```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
INPUT_FILE = "../../data/temperature/regional_temp_data.csv"
OUTPUT_FILE = "../../plots/temp_plot.png"
```

```python
df = pd.read_csv(INPUT_FILE, index_col = 0)
df.index = pd.to_datetime(df.index)
df
```

```python
fig, axs = plt.subplots(2, 5)
fig.set_figwidth(15)
fig.subplots_adjust(top=0.85)
plt.subplots_adjust(hspace=0.6)
for region in range(1, 11):
    i = (region-1) // 5
    j = (region-1) % 5
    axs[i, j].title.set_text(f"Region {region}")
    axs[i, j].plot(df.index, df[f"Region {region}"], linewidth = 0.75)
    axs[i, j].tick_params(axis = "x", labelrotation=45)
    axs[i, j].set_ylim(-25, 45)
    # if j != 0:
    #     axs[i, j].tick_params(labelleft = False)
fig.suptitle("Temperature (°C)", fontsize = 16)
fig.tight_layout()
fig.savefig(OUTPUT_FILE, dpi = 500)
```
