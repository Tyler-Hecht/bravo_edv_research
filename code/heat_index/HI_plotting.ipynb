```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"..\..\data\heat_index\cdc_regional_daily_heat_index.csv")
df["Date"] = pd.to_datetime(df["Date"])

fig, axes = plt.subplots(2, 5, figsize=(20, 8), sharex=True, sharey=True)
axes = axes.flatten()

for i, region in enumerate(sorted(df["HHS Region"].unique())):
    ax = axes[i]
    region_data = df[df["HHS Region"] == region]
    ax.plot(region_data["Date"], region_data["Region_Weighted_Heat_Index"], label=f"Region {region}")
    ax.set_title(f"HHS Region {region}")
    ax.set_xlabel("Date")
    ax.set_ylabel("Heat Index")

plt.tight_layout()
plt.show()
```
