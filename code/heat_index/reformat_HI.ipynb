```python
import pandas as pd
```

```python
INPUT_FILE = "../../data/heat_index/cdc_regional_daily_heat_index.csv"
OUTPUT_FILE = "../../data/heat_index/regional_heat_index_data.csv"
```

```python
df = pd.read_csv(INPUT_FILE)
df = df.pivot(columns = ["HHS Region"], index = "Date", values = "Region_Weighted_Heat_Index")
df.index.name = None
df.columns.name = None
df.rename(columns = lambda x: f"Region {x}")
df.to_csv(OUTPUT_FILE)
```
